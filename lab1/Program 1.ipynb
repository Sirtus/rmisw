Algorytm rekurencyjego mnożenia macierzy Binéta oparty jest na podziale dwóch mnożonych macierzy $A$ oraz $B$ na mniejsze podmacierze i wywołaniu tej samej operacji mnożenia dla uzyskanych podmacierzy. Wywoływanie kończy się gdy macierze mają rozmiar 2x2 i wykonywana jest standardowa operacja mnożenia macierzy 2x2. 

$AB = 
\begin{bmatrix}
A_{11}  &  A_{12} \\
A_{21}  &  A_{22} 
\end{bmatrix}
\begin{bmatrix}
B_{11}  &  B_{12} \\
B_{21}  &  B_{22} 
\end{bmatrix} =
\begin{bmatrix}
A_{11} B_{11}+ A_{12} B_{21}  & A_{11} B_{12} + A_{12} B_{22}\\
A_{21} B_{11} + A_{22} B_{21}  & A_{21} B_{12} + A_{22} B_{22}
\end{bmatrix}$

Algorytm można wyrazić poprzez pseudokod:

```
// input: A, B
// output: C
function binete_algorithm(A, B, C) {
    
    if (A.size > 2) {
        a11, a12, a21, a22 = A[range_A].slice_into_quarters();
        b11, b12, b21, b22 = B[range_B].slice_into_quarters();
        c11, c12, c21, c22 = C[range_C].slice_into_quarters();
        
        // up-left
        binete_algorithm(a11, b11, c11);
        binete_algorithm(a12, b21, c11);

        // up-right
        binete_algorithm(a11, b12, c12);
        binete_algorithm(a12, b22, c12);

        // down-left
        binete_algorithm(a21, b11, c21);
        binete_algorithm(a22, b21, c21);

        // down-right
        binete_algorithm(a21, b12, c22);
        binete_algorithm(a22, b22, c22);
    } else {
        mat_mul(A, B, C); // do traditional matrix multiplication
    }
}
```

````python
#binet
time_array_b = [0.000005, 0.000008, 0.000068, 0.000375, 0.003259, 0.024581, 0.185258, 1.431363]
add_b_arr =  [32, 256, 2048, 16384, 131072, 1048576, 8388608, 67108864]
mult_b_arr = [64, 512, 4096, 32768, 262144, 2097152, 16777216, 134217728]

#traditional
time_array_t = [0.000003, 0.000006, 0.000033, 0.000255, 0.001109, 0.009056, 0.083582, 0.735571]
add_t_arr =  [32, 256, 2048, 31744, 258048, 2080768, 16711680, 133955584]
mult_t_arr = [64, 512, 4096, 32768, 262144, 2097152, 16777216, 134217728]

l_arr = [i for i in range(2, 10)]
````

# $k$ = 9

````python
#binet
l_b = l_arr
time_b = time_array_b
add_b = add_b_arr
mult_b = mult_b_arr

#traditional
l_t = l_arr
time_t = time_array_t
add_t =  add_t_arr
mult_t = mult_t_arr
````

````python
import matplotlib.pyplot as plt

plt.figure(figsize=(20,10))
plt.plot(l_b, time_b, l_t, time_t)
plt.xlabel('Matrix size [2^N]')
plt.ylabel('Time[s]')
plt.legend(['Binét algorithm', 'Traditional multiplication'])
plt.xticks(range(l_t[0],l_t[len(l_t)-1]+1))
plt.show()
````

````python
plt.figure(figsize=(20,10))
plt.plot(l_b, add_b, l_t, add_t)
plt.xlabel('Matrix size [2^N]')
plt.ylabel('Number of floating point additions')
plt.legend(['Binét algorithm', 'Traditional multiplication'])
plt.xticks(range(l_t[0],l_t[len(l_t)-1]+1))
plt.show()
````

````python
plt.figure(figsize=(20,10))
plt.plot(l_b, mult_b, l_t, mult_t)
plt.xlabel('Matrix size [2^N]')
plt.ylabel('Number of floating point multiplications')
plt.legend(['Binét algorithm', 'Traditional multiplication'])
plt.xticks(range(l_t[0],l_t[len(l_t)-1]+1))
plt.show()
````
